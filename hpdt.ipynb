```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import csv
import re
import numpy as np
```

```python
r = requests.get('https://www.goodreads.com/list/show/1324.Top_100_New_Classics')
soup = bs(r.text, 'html.parser')

details=[]
series=[]
author=[]
avg_rating=[]
no_ratings=[]
score=[]
no_votes=[]
links=[]
description=[]
pages = []
year_published = []
publisher=[]
awards=[]
lang=[]
t=[]

for i in soup.findAll('span',{'itemprop':'name'}):
    details.append(i.get_text())
for i in soup.findAll('span',{'class':'minirating'}):
    avg_rating.append(re.findall("\d+\.\d+",i.get_text()))
for i in soup.findAll('span',{'class':'minirating'}):
    avg_rating.append(re.findall("\d+\.\d+",i.get_text())[0])
    no_ratings.append(re.findall(r'[\d,]+[,\d]',i.get_text())[1])

for i in soup.findAll('a',{'class':'bookTitle'}):
    links.append('https://goodreads.com'+i.get('href'))

title = details[::2]
author = details[1::2]

for i in title:
    txt = re.findall(r'\(.*?\)', i)
    if txt:
        series.append(txt[0].split(',')[0][1:])
    else:
        series.append('')
  




links
```

```python
def scrapper(w,d,ls):
    fi = soup.findAll(w,d)
    if fi:
        for i in fi: 
            ls.append(i.get_text())
    else:
        ls.append('')

for link in links:
    
    
    rs2 = requests.get(link)
    soup = bs(rs2.text, 'html.parser')
    
    
    scrapper('span',{'itemprop':'numberOfPages'},pages)
    scrapper('div',{'itemprop':'awards'},awards)
    scrapper('div',{'itemprop':'inLanguage'},lang)
    scrapper('div',{'id':"description"},description)
    scrapper('div',{'id':'details'},t)
```

```python
desc = [i.replace('\n','').replace('...more','') for i in description]
len(description)
```

```python
dt = pd.DataFrame({'title':title,'series':series,'author':author,'pages':pages,'avg_rating':avg_rating[:100],'no_ratings':no_ratings, 'description':desc})
```
